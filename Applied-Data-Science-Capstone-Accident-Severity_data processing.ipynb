```python
print("number of NaN values for the column UNDERINFL :", df['UNDERINFL'].isnull().sum())
print("number of NaN values for the column ROADCOND :", df['ROADCOND'].isnull().sum())
print("number of NaN values for the column LIGHTCOND :", df['LIGHTCOND'].isnull().sum())

print("number of NaN values for the column WEATHER :", df['WEATHER'].isnull().sum())

print("number of NaN values for the column SEVERITYCODE:", df['SEVERITYCODE'].isnull().sum())

```

```python
# use the non null data
df_main = df[["WEATHER","LIGHTCOND","ROADCOND","SPEEDING","SEVERITYCODE"]]
df_main.head()
```

```python
#Encoding Speeding(0 = No, 1 = Yes)
df_main["SPEEDING"].replace("Y", 1, inplace=True)
df_main["SPEEDING"].replace(np.nan, 0, inplace=True)
df_main.head()
```

```python
df_main1= df_main.dropna()
df_main1.head()
```

```python

print("number of NaN values for the column ROADCOND :", df_main1['ROADCOND'].isnull().sum())
print("number of NaN values for the column LIGHTCOND :", df_main1['LIGHTCOND'].isnull().sum())

print("number of NaN values for the column WEATHER :", df_main1['WEATHER'].isnull().sum())
print("number of NaN values for the column SEVERITYCODE:", df_main1['SEVERITYCODE'].isnull().sum())
```

```python
df_main1.shape
```

# Encoding

```python
#Encoding Light Conditions(0 = Light, 1 = Medium, 2 = Dark)
df_main1["LIGHTCOND"].replace("Daylight", 0, inplace=True)
df_main1["LIGHTCOND"].replace("Dark - Street Lights On", 1, inplace=True)
df_main1["LIGHTCOND"].replace("Dark - No Street Lights", 2, inplace=True)
df_main1["LIGHTCOND"].replace("Dusk", 1, inplace=True)
df_main1["LIGHTCOND"].replace("Dawn", 1, inplace=True)
df_main1["LIGHTCOND"].replace("Dark - Street Lights Off", 2, inplace=True)
df_main1["LIGHTCOND"].replace("Dark - Unknown Lighting", 2, inplace=True)
df_main1["LIGHTCOND"].replace("Other","Unknown", inplace=True)
df_main1.head()
```

```python
#Encoding Weather Conditions(0 = Clear, 1 = Overcast and Cloudy, 2 = Windy, 3 = Rain and Snow
df_main1["WEATHER"].replace("Clear", 0, inplace=True)
df_main1["WEATHER"].replace("Raining", 3, inplace=True)
df_main1["WEATHER"].replace("Overcast", 1, inplace=True)
df_main1["WEATHER"].replace("Other", "Unknown", inplace=True)
df_main1["WEATHER"].replace("Snowing", 3, inplace=True)
df_main1["WEATHER"].replace("Fog/Smog/Smoke", 2, inplace=True)
df_main1["WEATHER"].replace("Sleet/Hail/Freezing Rain", 3, inplace=True)
df_main1["WEATHER"].replace("Blowing Sand/Dirt", 2, inplace=True)
df_main1["WEATHER"].replace("Severe Crosswind", 2, inplace=True)
df_main1["WEATHER"].replace("Partly Cloudy", 1, inplace=True)
df_main1.head()
```

```python
#Encoding Road Conditions(0 = Dry, 1 = Mushy, 2 = Wet)
df_main1["ROADCOND"].replace("Dry", 0, inplace=True)
df_main1["ROADCOND"].replace("Wet", 2, inplace=True)
df_main1["ROADCOND"].replace("Ice", 2, inplace=True)
df_main1["ROADCOND"].replace("Snow/Slush", 1, inplace=True)
df_main1["ROADCOND"].replace("Other", "Unknown", inplace=True)
df_main1["ROADCOND"].replace("Standing Water", 2, inplace=True)
df_main1["ROADCOND"].replace("Sand/Mud/Dirt", 1, inplace=True)
df_main1["ROADCOND"].replace("Oil", 2, inplace=True)
df_main1.head()
```

```python
#Converting Speeding to int
df_main1["SPEEDING"]=df_main1["SPEEDING"].astype(int)
df_main1.head()
```
